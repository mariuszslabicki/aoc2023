```python
from functools import cache
import math

# 867270082910 is too low
# 884679773080 is too low

rec_counter = 0
total = 0
mem_dict = dict()

# @cache
def countOptions(field, groups, pos, group_no):
    global rec_counter
    global mem_dict
    global total
    rec_counter += 1
    if group_no == len(groups):
        if '#' not in field[pos:]:
            total += 1
            return
        else:
            return
    if pos >= len(field):
        return
    if field[pos] in ['.', '?']:
        next_pos = pos+1
        if next_pos <= len(field)-1:
            if (field, groups, next_pos, group_no) not in mem_dict:
                total_before = total
                countOptions(field, groups, next_pos, group_no)
                total_after = total
                mem_dict[(field, groups, next_pos, group_no)] = total_after-total_before
            else:
                total += mem_dict[(field, groups, next_pos, group_no)]
    
    chars_to_insert = groups[group_no]
    if field[pos] in ['?','#']:    
        fit_here = True
        if pos + chars_to_insert > len(field):
            fit_here = False

        if pos + chars_to_insert < len(field)-1 and field[pos + chars_to_insert] not in ['.', '?']:
            fit_here = False

        if fit_here is True:
            for i in range(chars_to_insert):
                if field[pos + i] not in ['?','#']:
                    fit_here = False

        if fit_here is True:
            par_a = pos+chars_to_insert+1
            par_b = group_no+1
            if (field, groups, par_a, par_b) not in mem_dict:
                total_before = total
                countOptions(field, groups, par_a, par_b)
                total_after = total
                mem_dict[(field, groups, par_a, par_b)] = total_after-total_before
            else:
                total += mem_dict[(field, groups, par_a, par_b)]

fields = []
groups_list = []

with open('input') as f:
    for line in f:
        line = line.strip()
        field = line.split()[0]
        for i in range(4):
            field += '?'+line.split()[0]
        field += '.'
        fields.append(tuple(field))
        second_part = line.split()[1].split(",")
        part_group = []
        for element in second_part:
            part_group.append(int((element)))
        group = []
        for i in range(5):
            group += part_group
        groups_list.append(tuple(group))

sum_total = 0

for field, groups in zip(fields, groups_list):
    mem_dict.clear()
    total = 0
    countOptions(field, groups, 0, 0)
    result = total
    sum_total += result

print(sum_total)
```
