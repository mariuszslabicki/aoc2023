```python
# %%timeit

# 5458 is too low
# 6094 is too low

import copy

map = set()
slopes = dict()

max_x = -1
max_y =  -1

start_pos = None
end_pos = None

with open('example_input') as f:
    y = 0
    for line in f:
        line = line.strip()
        if start_pos is None:
            start_pos = (line.index('.'), 0)
        max_x = len(line)
        for x in range(len(line)):
            if line[x] != '#':
                map.add((x, y))
            if line[x] in ['<', '>', 'v', '^']:
                slopes[(x, y)] = line[x]
        end_pos = (line.index('.'), y)
        y += 1
        max_y = y

nodes = [start_pos, end_pos]

for y in range(max_y):
    for x in range(max_x):
        if (x, y) not in map:
            continue
        shifts = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        exits = 0
        for shift in shifts:
            if (x + shift[0], y + shift[1]) in map:
                exits += 1
        if exits > 2:
            nodes.append((x, y))

vertices = dict()


for node in nodes:
    paths = [[node]]
    found_paths = []
    while len(paths) > 0:
        new_paths = []
        for path in paths:
            if len(path) > 1 and path[-1] in nodes:
                found_paths.append([len(path)-1, path[-1]])
                continue
            shifts = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            for shift in shifts:
                new_candidate = (path[-1][0] + shift[0], path[-1][1] + shift[1])
                if new_candidate not in map or new_candidate in path:
                    continue
                new_path = path.copy()
                new_path.append(new_candidate)
                new_paths.append(new_path)
        paths = new_paths
    vertices[node] = found_paths

for node in nodes:
    for step_a in vertices[node]:
        if node == step_a[1]:
            continue
        for step_b in vertices[step_a[1]]:
            if step_a == step_b[1]:
                continue
            for step_c in vertices[step_b[1]]:
                if step_b == step_c[1]:
                    continue
                if node == step_c[1]:
                    print("Mam koleczko", node, step_a[1], step_b[1], step_c[1])

paths = [[0, set(), start_pos]]
max_found = -1
while len(paths) > 0:
    new_paths = []
    for path in paths:
        if path[2] == end_pos:
            if path[0] > max_found:
                max_found = path[0]
                print("Dojechalem do konca", path[0], "max:", max_found)
        for exit_link in vertices[path[2]]:
            if exit_link[1] in path[1]:
                continue
            new_length = path[0] + exit_link[0]
            new_path = copy.deepcopy(path)
            new_path[0] = new_length
            new_path[1].add(new_path[2])
            new_path[2] = exit_link[1]
            new_paths.append(new_path)
    paths = new_paths
print(max_found)
```
