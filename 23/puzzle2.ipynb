```python
# %%timeit

# 5458 is too low
# 6094 is too low

import copy

map = set()
slopes = dict()

max_x = -1
max_y =  -1

start_pos = None
end_pos = None

with open('input') as f:
    y = 0
    for line in f:
        line = line.strip()
        if start_pos is None:
            start_pos = (line.index('.'), 0)
        max_x = len(line)
        for x in range(len(line)):
            if line[x] != '#':
                map.add((x, y))
            if line[x] in ['<', '>', 'v', '^']:
                slopes[(x, y)] = line[x]
        end_pos = (line.index('.'), y)
        y += 1
        max_y = y

nodes = [start_pos, end_pos]

for y in range(max_y):
    for x in range(max_x):
        if (x, y) not in map:
            continue
        shifts = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        exits = 0
        for shift in shifts:
            if (x + shift[0], y + shift[1]) in map:
                exits += 1
        if exits > 2:
            nodes.append((x, y))

vertices = dict()


for node in nodes:
    paths = [[node]]
    found_paths = []
    while len(paths) > 0:
        new_paths = []
        for path in paths:
            if len(path) > 1 and path[-1] in nodes:
                found_paths.append([len(path)-1, path[-1]])
                continue
            shifts = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            for shift in shifts:
                new_candidate = (path[-1][0] + shift[0], path[-1][1] + shift[1])
                if new_candidate not in map or new_candidate in path:
                    continue
                new_path = path.copy()
                new_path.append(new_candidate)
                new_paths.append(new_path)
        paths = new_paths
    vertices[node] = found_paths

removed = set()

start_search_again = True
while start_search_again is True:
    vert_to_remove = None
    start_search_again = False
    found_to_remove = False
    for node in nodes:
        for step_a in vertices[node]:
            if node == step_a[1] or step_a[1] in removed:
                continue
            for step_b in vertices[step_a[1]]:
                if step_a[1] == step_b[1] or step_b[1] in removed:
                    continue
                for step_c in vertices[step_b[1]]:
                    if step_b[1] == step_c[1] or step_c[1] in removed:
                        continue
                    if node == step_c[1]:
                        smallest_len = 99999999999
                        smallest_link = None
                        print("Cycle", node, step_a[1], step_b[1], step_c[1])
                        for exit in vertices[node]:
                            if exit[1] == step_a[1]:
                                if step_a[0] < smallest_len:
                                    smallest_len = step_a[0]
                                    smallest_link = [node, step_a[1]]
                        for exit in vertices[step_a[1]]:
                            if exit[1] == step_b[1]:
                                if step_b[0] < smallest_len:
                                    smallest_len = step_b[0]
                                    smallest_link = [step_a[1], step_b[1]]
                        for exit in vertices[step_c[1]]:
                            if exit[1] == step_c[1]:
                                if step_c[0] < smallest_len:
                                    smallest_len = step_c[0]
                                    smallest_link = [step_b[1], step_c[1]]
                        removed.add(smallest_link[0])
                        removed.add(smallest_link[1])
                        vert_to_remove = smallest_link
                        found_to_remove = True
                        start_search_again = True
                        if found_to_remove is True:
                            break
                    if found_to_remove is True:
                        break
                if found_to_remove is True:
                    break
            if found_to_remove is True:
                break
        if found_to_remove is True:
            break
    if found_to_remove is True:
        print("To remove", vert_to_remove) 
        removed.add(vert_to_remove[0])
        removed.add(vert_to_remove[1])
        for exit in vertices[vert_to_remove[0]]:
            if exit[1] == vert_to_remove[1]:
                entry_to_remove = exit
                break
        vertices[vert_to_remove[0]].remove(entry_to_remove)

        for exit in vertices[vert_to_remove[1]]:
            if exit[1] == vert_to_remove[0]:
                entry_to_remove = exit
                break
        vertices[vert_to_remove[1]].remove(entry_to_remove)

paths = [[0, set(), start_pos]]
max_found = -1
while len(paths) > 0:
    new_paths = []
    for path in paths:
        if path[2] == end_pos:
            if path[0] > max_found:
                max_found = path[0]
                print("Dojechalem do konca", path[0], "max:", max_found)
        for exit_link in vertices[path[2]]:
            if exit_link[1] in path[1]:
                continue
            new_length = path[0] + exit_link[0]
            new_path = copy.deepcopy(path)
            new_path[0] = new_length
            new_path[1].add(new_path[2])
            new_path[2] = exit_link[1]
            new_paths.append(new_path)
    paths = new_paths
print(max_found)
```
