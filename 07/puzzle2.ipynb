```python
import functools
from collections import Counter

bids = {}

hands = []

with open('input') as f:
    for line in f:
        hand = str(line.strip()).split(" ")[0]
        bid = int(str(line.strip()).split(" ")[1])
        bids[hand] = bid
        hands.append(hand)

def returnCardVal(card):
    if card not in ["T", "J", "Q", "K", "A"]:
        cardVal = int(card)
    else:
        if card == 'T':
            cardVal = 10
        if card == 'J':
            cardVal = 1
        if card == 'Q':
            cardVal = 12
        if card == 'K':
            cardVal = 13
        if card == 'A':
            cardVal = 14
    return cardVal

def returnHandVal(hand):
    cardNo = sorted(Counter(hand).values(), reverse=True)
    counter = Counter(hand)
    noOfJokers = counter['J']
    
    if noOfJokers == 5:
        return 80

    handUpdated = [card for card in hand if card != 'J']
    cardNo = sorted(Counter(handUpdated).values(), reverse=True)

    if cardNo[0] + noOfJokers >= 5:
        return 80
    if cardNo[0] + noOfJokers == 4:
        return 70
    if cardNo[0] + noOfJokers == 3 and cardNo[1] == 2:
        return 60
    if cardNo[0] + noOfJokers == 3 and cardNo[1] == 1:
        return 50
    if cardNo[0] + noOfJokers == 2 and cardNo[1] == 2:
        return 40
    if cardNo[0] + noOfJokers == 2 and cardNo[1] == 1:
        return 30
    if cardNo[0] == 1:
        return 20

def compareHands(handA, handB):
    valA = returnHandVal(handA)
    valB = returnHandVal(handB)
    if valA > valB:
        return 1
    if valA < valB:
        return -1
    if valA == valB:
        for i in range(len(handA)):
            cardA = returnCardVal(handA[i])
            cardB = returnCardVal(handB[i])
            if cardA > cardB:
                return 1
            if cardA < cardB:
                return -1

hands = sorted(hands, key=functools.cmp_to_key(compareHands))

total = 0

for i in range(len(hands)):
    total += (i+1) * bids[hands[i]]

print(total)
```
