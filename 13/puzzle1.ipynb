```python
# 34645 is too low

def findReflection(matrix):
    result = 0
    #columns
    rows = dict()
    columns = dict()
    for x in range(len(matrix[0])):
        column_desc = []
        for y in range(len(matrix)):
            if matrix[y][x] == 1:
                column_desc.append(y)
        columns[x] = column_desc
        
    for x in range(len(matrix[0])-1):
        if columns[x] == columns[x+1]:
            l = x
            r = x+1
            found = True
            while l >= 0 and r <= len(matrix[0])-1:
                if columns[l] != columns[r]:
                    found = False
                    break
                l -= 1
                r += 1
            if found is True:
                result += x+1
                break

    #rows
    for y in range(len(matrix)):
        row_desc = []
        for x in range(len(matrix[0])):
            if matrix[y][x] == 1:
                row_desc.append(x)
        rows[y] = row_desc
        
    for y in range(len(matrix)-1):
        if rows[y] == rows[y+1]:
            u = y
            d = y+1
            found = True
            while u >= 0 and d <= len(matrix)-1:
                if rows[u] != rows[d]:
                    found = False
                    break
                u -= 1
                d += 1
            if found is True:
                result += 100*(y+1)
                break

    return result


total = 0

with open('example_input') as f:
    matrix = []
    row_no = 0
    for line in f:
        line = line.strip()
        if len(line) > 1:
            new_row = []
            for element in line:
                if element == '#':
                    new_row.append(1)
                else:
                    new_row.append(0)
            matrix.append(new_row)
        else:
            total += findReflection(matrix)
            matrix = []

    total += findReflection(matrix)

print(total)
```
