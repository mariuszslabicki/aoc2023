```python
# 53859 too high

values = []

with open('input') as f:
    current_scanner = []
    for line in f:
        reading = str(line.strip())
        for i in range(len(reading)):
            if reading[i].isnumeric():
                dec = 10*int(reading[i])
                break
            if reading[i:i+len("one")] == "one":
                dec = 10*1
                break
            if reading[i:i+len("two")] == "two":
                dec = 10*2
                break
            if reading[i:i+len("three")] == "three":
                dec = 10*3
                break
            if reading[i:i+len("four")] == "four":
                dec = 10*4
                break
            if reading[i:i+len("five")] == "five":
                dec = 10*5
                break
            if reading[i:i+len("six")] == "six":
                dec = 10*6
                break
            if reading[i:i+len("seven")] == "seven":
                dec = 10*7
                break
            if reading[i:i+len("eight")] == "eight":
                dec = 10*8
                break
            if reading[i:i+len("nine")] == "nine":
                dec = 10*9
                break
        reading = reading[::-1]
        for i in range(len(reading)):
            if reading[i].isnumeric():
                uni = int(reading[i])
                break
            if reading[i:i+len("eno")] == "eno":
                uni = 1
                break
            if reading[i:i+len("owt")] == "owt":
                uni = 2
                break
            if reading[i:i+len("eerht")] == "eerht":
                uni = 3
                break
            if reading[i:i+len("ruof")] == "ruof":
                uni = 4
                break
            if reading[i:i+len("evif")] == "evif":
                uni = 5
                break
            if reading[i:i+len("xis")] == "xis":
                uni = 6
                break
            if reading[i:i+len("neves")] == "neves":
                uni = 7
                break
            if reading[i:i+len("thgie")] == "thgie":
                uni = 8
                break
            if reading[i:i+len("enin")] == "enin":
                uni = 9
                break
        values.append(dec+uni)

print("Total:", sum(values))
```
