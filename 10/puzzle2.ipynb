```python
def fillZeros(x, y, max_x, max_y, fields):
    seen_fields = [[x, y]]
    edges = [[x, y]]

    while len(edges) > 0:
        new_edges = []
        for edge in edges:
            shifts = [[0, 1], [0, -1], [1, 0], [-1, 0]]
            for shift in shifts:
                new_edge = [edge[0] + shift[0], edge[1] + shift[1]]
                if new_edge[0] < 0:
                    continue
                if new_edge[0] >= max_x:
                    continue
                if new_edge[1] < 0:
                    continue
                if new_edge[1] >= max_y:
                    continue
                if tuple(new_edge) in fields or new_edge in seen_fields:
                    continue
                new_edges.append(new_edge)
                seen_fields.append(new_edge)
        edges = new_edges.copy()
    return seen_fields

fields = {}
start = None

max_x = 0
max_y = 0

with open('example_input') as f:
    row_no = 0
    for line in f:
        line = line.strip()
        for i in range(len(line)):
            if line[i] != '.':
                fields[(i, row_no)] = line[i]
                if line[i] == 'S':
                    start = (i, row_no)
        row_no += 1
    max_x = len(line)
    max_y = row_no
print(start)
paths = [[start]]

found_path = None

while True:
    new_paths = []
    found = False
    for path in paths:
        shifts = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        candidates = []
        for shift in shifts:
            new_field = (path[-1][0]+shift[0], path[-1][1]+shift[1])
            if new_field not in fields:
                continue
            if new_field in path and new_field != start:
                continue
            if shift == (1, 0): #right
                if fields[path[-1]] not in ['-', 'L', 'F', 'S']:
                    continue
                if fields[new_field] not in ['-', '7', 'J', 'S']:
                    continue
            if shift == (-1, 0): #left
                if fields[path[-1]] not in ['-', 'J', '7', 'S']:
                    continue
                if fields[new_field] not in ['-', 'L', 'F', 'S']:
                    continue
            if shift == (0, 1): #down
                if fields[path[-1]] not in ['|', 'F', '7', 'S']:
                    continue
                if fields[new_field] not in ['|', 'L', 'J', 'S']:
                    continue
            if shift == (0, -1): #up
                if fields[path[-1]] not in ['|', 'L', 'J', 'S']:
                    continue
                if fields[new_field] not in ['|', 'F', '7', 'S']:
                    continue
            if len(path) > 4 and new_field == start:
                found = True
                found_path = path.copy()
                print(len(path)//2)
                break
            if found is True:
                break
            new_path = path.copy()
            new_path.append(new_field)
            new_paths.append(new_path)
        if found is True:
            break
    if found is True:
        break
    paths = new_paths

print("Max x", max_x)
print("Max y", max_y)

extended_fields = {}

for x in range(max_x):
    for y in range(max_y):
        if (x, y) in fields:
            extended_fields[(2*x, 2*y)] = fields[(x, y)]
            if fields[(x, y)] in ['S', '-', 'L', 'F'] and (x+1, y) in fields and fields[(x+1, y)] in ['-', '7', 'J']:
                extended_fields[(2*x+1, 2*y)] = '-'
            if fields[(x, y)] in ['S', '|', '7', 'F'] and (x, y+1) in fields and fields[(x, y+1)] in ['|', 'L', 'J']:
                extended_fields[(2*x, 2*y+1)] = '|'

print("path", found_path)
print("ext fields len", len(extended_fields))
print("ext fields", extended_fields)


ext_zeros = []

ext_max_x = (max_x)*2
ext_max_y = (max_y)*2

for x in [0, ext_max_x-1]:
    for y in range(ext_max_y):
        if (x, y) not in extended_fields:
            if [x, y] not in ext_zeros:
                visited = fillZeros(x, y, ext_max_x-1, ext_max_y-1, extended_fields)
                for element in visited:
                    if element not in ext_zeros:
                        ext_zeros.append(element)

for x in range(ext_max_x):
    for y in [0, ext_max_y-1]:
        if (x, y) not in extended_fields:
            if [x, y] not in ext_zeros:
                visited = fillZeros(x, y, ext_max_x-1, ext_max_y-1, extended_fields)
                for element in visited:
                    if element not in ext_zeros:
                        ext_zeros.append(element)

inside = 0

for x in range(max_x):
    for y in range(max_y):
        if (2*x, 2*y) not in extended_fields and [2*x, 2*y] not in ext_zeros:
            inside += 1

print(inside)

for y in range(max_y):
    for x in range(max_x):
        if (x, y) in fields:
            print(fields[(x, y)], end="")
            continue
        if [2*x, 2*y] in ext_zeros:
            print('0', end="")
            continue
        print('I', end="")
    print()
```
