```python
# 1850633429 too high

seed_ranges = []

sts_r = False
stf_r = False
ftw_r = False
wtl_r = False
ltt_r = False
tth_r = False
htl_r = False

sts = []
stf = []
ftw = []
wtl = []
ltt = []
tth = []
htl = []

almanac = []

with open('example_input') as f:
    for line in f:
        line = str(line.strip()).split(" ")
        if line[0] == 'seeds:':
            for i in range(1, len(line), 2):
                seed_ranges.append([int(line[i]), int(line[i])+int(line[i+1])])

        if line[0] == 'seed-to-soil':
            sts_r = True
            continue
        if sts_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            sts.append(new_recipe)
        if sts_r is True and len(line) == 1:
            almanac.append(sts)
            sts_r = False

        if line[0] == 'soil-to-fertilizer':
            stf_r = True
            continue
        if stf_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            stf.append(new_recipe)
        if stf_r is True and len(line) == 1:
            almanac.append(stf)
            stf_r = False

        if line[0] == 'fertilizer-to-water':
            ftw_r = True
            continue
        if ftw_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            ftw.append(new_recipe)
        if ftw_r is True and len(line) == 1:
            almanac.append(ftw)
            ftw_r = False

        if line[0] == 'water-to-light':
            wtl_r = True
            continue
        if wtl_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            wtl.append(new_recipe)
        if wtl_r is True and len(line) == 1:
            almanac.append(wtl)
            wtl_r = False

        if line[0] == 'light-to-temperature':
            ltt_r = True
            continue
        if ltt_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            ltt.append(new_recipe)
        if ltt_r is True and len(line) == 1:
            almanac.append(ltt)
            ltt_r = False

        if line[0] == 'temperature-to-humidity':
            tth_r = True
            continue
        if tth_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            tth.append(new_recipe)
        if tth_r is True and len(line) == 1:
            almanac.append(tth)
            tth_r = False

        if line[0] == 'humidity-to-location':
            htl_r = True
            continue
        if htl_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            htl.append(new_recipe)
        if htl_r is True and len(line) == 1:
            almanac.append(htl)
            htl_r = False

    almanac.append(htl)

min_location = None

for seed_range in seed_ranges:
    for val in range(seed_range[0], seed_range[1]):
        for page in almanac:
            for recipe in page:
                if val >= recipe[0] and val <= recipe[1]:
                    val += recipe[2]
                    break
        if min_location is None or val < min_location:
            min_location = val
            print("Found min:", min_location)

print(min_location)
```
