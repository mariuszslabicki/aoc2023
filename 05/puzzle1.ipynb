```python
seeds = []

sts_r = False
stf_r = False
ftw_r = False
wtl_r = False
ltt_r = False
tth_r = False
htl_r = False

sts = []
stf = []
ftw = []
wtl = []
ltt = []
tth = []
htl = []

with open('input') as f:
    for line in f:
        line = str(line.strip()).split(" ")
        if line[0] == 'seeds:':
            for element in line[1:]:
                seeds.append(int(element))

        if line[0] == 'seed-to-soil':
            sts_r = True
            continue
        if sts_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            sts.append(new_recipe)
        if sts_r is True and len(line) == 1:
            sts_r = False

        if line[0] == 'soil-to-fertilizer':
            stf_r = True
            continue
        if stf_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            stf.append(new_recipe)
        if stf_r is True and len(line) == 1:
            stf_r = False

        if line[0] == 'fertilizer-to-water':
            ftw_r = True
            continue
        if ftw_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            ftw.append(new_recipe)
        if ftw_r is True and len(line) == 1:
            ftw_r = False

        if line[0] == 'water-to-light':
            wtl_r = True
            continue
        if wtl_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            wtl.append(new_recipe)
        if wtl_r is True and len(line) == 1:
            wtl_r = False

        if line[0] == 'light-to-temperature':
            ltt_r = True
            continue
        if ltt_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            ltt.append(new_recipe)
        if ltt_r is True and len(line) == 1:
            ltt_r = False

        if line[0] == 'temperature-to-humidity':
            tth_r = True
            continue
        if tth_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            tth.append(new_recipe)
        if tth_r is True and len(line) == 1:
            tth_r = False

        if line[0] == 'humidity-to-location':
            htl_r = True
            continue
        if htl_r is True and len(line) > 1:
            new_recipe = [int(line[1]), int(line[1]) + int(line[2])-1, int(line[0]) - int(line[1])]
            htl.append(new_recipe)
        if htl_r is True and len(line) == 1:
            htl_r = False

locations = []

for seed in seeds:
    for recipe in sts:
        soil = seed
        if seed >= recipe[0] and seed <= recipe[1]:
            soil += recipe[2]
            break

    for recipe in stf:
        fertilizer = soil
        if soil >= recipe[0] and soil <= recipe[1]:
            fertilizer += recipe[2]
            break
    
    for recipe in ftw:
        water = fertilizer
        if fertilizer >= recipe[0] and fertilizer <= recipe[1]:
            water += recipe[2]
            break

    for recipe in wtl:
        light = water
        if water >= recipe[0] and water <= recipe[1]:
            light += recipe[2]
            break

    for recipe in ltt:
        temp = light
        if light >= recipe[0] and light <= recipe[1]:
            temp += recipe[2]
            break

    for recipe in tth:
        hum = temp
        if temp >= recipe[0] and temp <= recipe[1]:
            hum += recipe[2]
            break

    for recipe in htl:
        loc = hum
        if hum >= recipe[0] and hum <= recipe[1]:
            loc += recipe[2]
            break
    locations.append(loc)


print(min(locations))
```
