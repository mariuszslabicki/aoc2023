```python
losses = dict()

max_x = -1
max_y =  -1

with open('example_input') as f:
    y = 0
    for line in f:
        line = line.strip()
        print(line)
        max_x = len(line)
        for x in range(len(line)):
            losses[(x, y)] = int(line[x])
        y += 1
        max_y = y

cur_positions = []
cur_positions.append([-1, 0, -losses[(0, 0)]])

smallest_sum = dict()

shift = [0, 1]

print(max_x, max_y)

step = 0

shortest_path = None
move_range = 5

while len(cur_positions) > 0:
    new_positions = []
    if shift == [0, 1]:
        shift = [1, 0]
    else:
        shift = [0, 1]
    # print("Step")
    for position in cur_positions:
        # print(position)
        latest_sum = position[2]
        for i in range(1, move_range):
            new_x = position[0]+i*shift[0]
            new_y = position[1]+i*shift[1]
            if new_x >= 0 and new_x < max_x and new_y >= 0 and new_y < max_y:
                new_sum = latest_sum + losses[(new_x, new_y)]
                if shortest_path is not None and new_sum > shortest_path:
                    continue
                new_position = [new_x, new_y, new_sum]
                latest_sum = new_sum
                if (new_x, new_y) not in smallest_sum:
                    smallest_sum[(new_x, new_y)] = new_sum
                    new_positions.append(new_position)
                else:
                    if smallest_sum[(new_x, new_y)] > new_sum:
                        smallest_sum[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
        for i in range(-1, -4, -1):
            new_x = position[0]+i*shift[0]
            new_y = position[1]+i*shift[1]
            if new_x >= 0 and new_x < max_x and new_y >= 0 and new_y < max_y:
                new_sum = latest_sum + losses[(new_x, new_y)]
                if shortest_path is not None and new_sum > shortest_path:
                    continue
                new_position = [new_x, new_y, new_sum]
                latest_sum = new_sum
                if (new_x, new_y) not in smallest_sum:
                    smallest_sum[(new_x, new_y)] = new_sum
                    new_positions.append(new_position)
                else:
                    if smallest_sum[(new_x, new_y)] > new_sum:
                        smallest_sum[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
    if move_range == 5:
        new_positions.pop(0)
    move_range = 4
    # print(new_positions)
    # if step == 4:
    #     break
    step += 1
    # break
    cur_positions = new_positions
    for position in cur_positions:
        if position[0] == max_x-1 and position[1] == max_y-1:
            shortest_path = position[2]
            print("znalazlem", position)
            print(position[2])
            break
```
