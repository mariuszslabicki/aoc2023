```python
losses = dict()

max_x = -1
max_y =  -1

with open('example_input') as f:
    y = 0
    for line in f:
        line = line.strip()
        print(line)
        max_x = len(line)
        for x in range(len(line)):
            losses[(x, y)] = int(line[x])
        y += 1
        max_y = y

cur_positions = []
cur_positions.append([0, 0, 0])

smallest_sum = dict()

shift = [0, 1]

print(max_x, max_y)

step = 0

while True:
    print("step", step)
    print(len(smallest_sum))
    new_positions = []
    if shift == [0, 1]:
        shift = [1, 0]
    else:
        shift = [0, 1]
    for position in cur_positions:
        for i in range(-3, 4, 1):
            if i == 0:
                continue
            new_x = position[0]+i*shift[0]
            new_y = position[1]+i*shift[1]
            if new_x >= 0 and new_x < max_x and new_y >= 0 and new_y < max_y:
                new_sum = position[2] + losses[(new_x, new_y)]
                new_position = [new_x, new_y, new_sum]
                if (new_x, new_y) not in smallest_sum:
                    smallest_sum[(new_x, new_y)] = new_sum
                    new_positions.append(new_position)
                else:
                    if smallest_sum[(new_x, new_y)] > new_sum:
                        smallest_sum[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
                
    print(new_positions)
    if step == 4:
        break
    step += 1
    # break
    cur_positions = new_positions
    for position in cur_positions:
        if position[0] == max_x-1 and position[1] == max_y-1:
            print("znalazlem", position)
            print(position[2])
            break
```
