```python
losses = dict()

max_x = -1
max_y =  -1

with open('input') as f:
    y = 0
    for line in f:
        line = line.strip()
        # print(line)
        max_x = len(line)
        for x in range(len(line)):
            losses[(x, y)] = int(line[x])
        y += 1
        max_y = y

cur_positions = []
cur_positions.append([0, 0, 0])

shift = [0, 1]

step = 0

vertical_starts = dict()
horizontal_starts = dict()

shortest_path = None

while len(cur_positions) > 0:
    new_positions = []
    if shift == [0, 1]:
        shift = [1, 0]
    else:
        shift = [0, 1]
    # print("Step")
    for position in cur_positions:
        # print(position)
        latest_sum = position[2]
        for i in range(1, 4):
            new_x = position[0]+i*shift[0]
            new_y = position[1]+i*shift[1]
            if new_x >= 0 and new_x < max_x and new_y >= 0 and new_y < max_y:
                new_sum = latest_sum + losses[(new_x, new_y)]
                if shortest_path is not None and new_sum > shortest_path:
                    continue
                new_position = [new_x, new_y, new_sum]
                latest_sum = new_sum
                # new_positions.append(new_position)
                if shift == [1, 0]:
                    if (new_x, new_y) not in horizontal_starts:
                        horizontal_starts[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
                    else:
                        if horizontal_starts[(new_x, new_y)] > new_sum:
                            horizontal_starts[(new_x, new_y)] = new_sum
                            new_positions.append(new_position)

                if shift == [0, 1]:
                    if (new_x, new_y) not in vertical_starts:
                        vertical_starts[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
                    else:
                        if vertical_starts[(new_x, new_y)] > new_sum:
                            vertical_starts[(new_x, new_y)] = new_sum
                            new_positions.append(new_position)
        latest_sum = position[2]
        for i in range(-1, -4, -1):
            new_x = position[0]+i*shift[0]
            new_y = position[1]+i*shift[1]
            if new_x >= 0 and new_x < max_x and new_y >= 0 and new_y < max_y:
                new_sum = latest_sum + losses[(new_x, new_y)]
                if shortest_path is not None and new_sum > shortest_path:
                    continue
                new_position = [new_x, new_y, new_sum]
                latest_sum = new_sum
                # new_positions.append(new_position)
                if shift == [1, 0]:
                    if (new_x, new_y) not in horizontal_starts:
                        horizontal_starts[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
                    else:
                        if horizontal_starts[(new_x, new_y)] > new_sum:
                            horizontal_starts[(new_x, new_y)] = new_sum
                            new_positions.append(new_position)

                if shift == [0, 1]:
                    if (new_x, new_y) not in vertical_starts:
                        vertical_starts[(new_x, new_y)] = new_sum
                        new_positions.append(new_position)
                    else:
                        if vertical_starts[(new_x, new_y)] > new_sum:
                            vertical_starts[(new_x, new_y)] = new_sum
                            new_positions.append(new_position)

    cur_positions = new_positions
    for position in cur_positions:
        if position[0] == max_x-1 and position[1] == max_y-1:
            if shortest_path is None or shortest_path > position[2]:
                shortest_path = position[2]
                print("znalazlem", position)
                print(shortest_path)

print(shortest_path)
```
